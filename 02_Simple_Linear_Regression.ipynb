### Import Libraries

```python
import joblib
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

### Load Data

```python
X_train = joblib.load('Data/X_train.pkl')
X_test = joblib.load('Data/X_test.pkl')
Y_train = joblib.load('Data/Y_train.pkl')
Y_test = joblib.load('Data/Y_test.pkl')
```

### Explore Data

```python
X_train.info()
```

```python
X_train.head(10)
```

### Selecting Feature

```python
X = X_train[['lstat']]
x = X_test[['lstat']]
```

### Train Model on feature: lstat

```python
model = LinearRegression()
model.fit(X, Y_train)
```

### Predict median-value(in 1000$) for a given lstat in testing data

```python
y_pred = model.predict(x)
```

#### Evaluates the model

```python
model.score(x, Y_test)*100
```

##### Estimated coefficient for the linear regression problem

```python
model.coef_
```

#### Independent term in the linear model 

```python
model.intercept_
```

#### Number of feature seen during the fit

```python
model.n_features_in_
```

### Visualize the model

```python
plt.scatter(x, Y_test, color="black")
plt.plot(x, y_pred, color='blue', linewidth=3)
```
